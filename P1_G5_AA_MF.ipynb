#Modelo final


```python
import pandas as pd
import joblib

# Cargamos nuestro modelo final
modelo_cargado = joblib.load('modelo_final.pkl')

# Cargamos nuestros datos de competición
df_competition = pd.read_pickle('bank_competition_05.pkl')

# Realizamos las predicciones
predicciones_competicion = modelo_cargado.predict(df_competition)

# Guardamos nuestras predicciones en un fichero CSV
df_predicciones = pd.DataFrame({'Prediccion': predicciones_competicion})
df_predicciones.to_csv('predicciones_G5.csv', index=False)
```
